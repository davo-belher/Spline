```python
import numpy as np
import matplotlib.pyplot as plt


class spline():
    
    def knots(self, x, y):
        self.X = x
        self.Y = y
        print('valores en x: \n', x)
        print('valores en y: \n', y)
        vx = np.concatenate(([0],x,[0]))
        vy = np.concatenate(([0],y,[0]))
        print('vector de x: \n',vx)
        print('vector de y; \n',vy)
        self.vecx = vx
        self.vecy = vy
        
    def interpolate(self):
        n = len(self.X)
        self.n = n
        m0 = np.zeros((n+2,n+2))
        print(m0)
        for i in range(n):
            m0[i+1,i]=1
            m0[i+1,i+1]=4
            m0[i+1,i+2]=1            
        m0[0,0]=1
        m0[0,1]=-2
        m0[0,2]=1
        m0[n+1,n-1]=1
        m0[n+1,n]=-2
        m0[n+1,n+1]=1
        print('matriz de coeficientes: \n', m0)
        mI0 = np.linalg.inv(m0)
        print('matriz inverza de coeficientes: \n', mI0)
        self.mat0 = m0
        self.matInv0 = mI0
        
    def contorlPoints(self):
        mI0 = self.matInv0
        cx = 6 * mI0.dot(self.vecx)
        cy = 6 * mI0.dot(self.vecy)
        print('valor de puntos de control en x: \n', cx)
        print('valor de puntos de control en y: \n', cy)
        self.ctrx = cx
        self.ctry = cy
    
    def grafSpline(self):
        t=np.linspace(0,1,num=101)
        cx = self.ctrx
        cy = self.ctry
        for i in range(0,self.n-1):
            xt = (1.0/6.0)*(cx[i]*((1 - t)**3) + cx[i+1]*(4 - 6*t**2 + 3*t**3) + cx[i+2]*(1 + 3*t + 3*t**2 - 3*t**3) + cx[i+3]*(t**3))
            yt = (1.0/6.0)*(cy[i]*((1 - t)**3) + cy[i+1]*(4 - 6*t**2 + 3*t**3) + cy[i+2]*(1 + 3*t + 3*t**2 - 3*t**3) + cy[i+3]*(t**3))
            plt.plot(xt,yt,color = 'gray')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()
            
    def grafSplinePoint(self):
        cx = self.ctrx
        cy = self.ctry
        plt.plot(cx,cy, markersize = 5, label = 'Control Points' )
        plt.xlabel('x')
        plt.ylabel('y')
        plt.legend(loc='lower right', ncol=1)
        plt.show()
        
    def grafSplineKnots(self):
        x = self.X
        y = self.Y
        plt.plot(x,y, markersize = 5, label = 'Knots')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.legend(loc='lower right', ncol=1)
        plt.show()
        
    def grafSplineAll(self):
        t=np.linspace(0,1,num=101)
        cx = self.ctrx
        cy = self.ctry
        x = self.X
        y = self.Y
        for i in range(0,self.n-1):
            xt = (1.0/6.0)*(cx[i]*((1 - t)**3) + cx[i+1]*(4 - 6*t**2 + 3*t**3) + cx[i+2]*(1 + 3*t + 3*t**2 - 3*t**3) + cx[i+3]*(t**3))
            yt = (1.0/6.0)*(cy[i]*((1 - t)**3) + cy[i+1]*(4 - 6*t**2 + 3*t**3) + cy[i+2]*(1 + 3*t + 3*t**2 - 3*t**3) + cy[i+3]*(t**3))
            plt.plot(xt , yt, color = 'gray')
        plt.plot(cx ,cy , markersize = 5, label = 'Control Points')
        plt.plot(x , y, markersize = 5, label = 'Knots')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.legend(loc='lower right', ncol=1)
        plt.show()
        
def main():  
    
    SP = spline()
    SP.knots(CIX, CDX)
    SP.interpolate()
    SP.contorlPoints()
    SP.grafSpline()
    SP.grafSplinePoint()
    SP.grafSplineKnots()
    SP.grafSplineAll()
    
if __name__ == "__main__":
    main()
```
